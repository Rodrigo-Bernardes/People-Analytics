# 🔄📄🥼 <span style="color: white; background-color: Teal"><b> Atualização do Power BI de Atestados Médicos </b></span></p>

📈 <span style="color: MediumSeaGreen"><b> 1- Abertura automática do relatório Power BI Desktop </b></span></p>
- O script inicia carregando o arquivo .pbix localizado em um diretório de rede da área de Gestão de Pessoas
- Ele usa:
    - os.startfile() para abrir o PBIX  
    - temporizadores (time.sleep) calibrados para aguardar o carregamento completo da aplicação
- Isso garante que o Power BI desktop esteja totalmente operacional antes de qualquer interação

🚢 <span style="color: MediumSeaGreen"><b> 2- Navegação automatizada dentro do Power BI </b></span></p>
- Após abrir o PBIX, o fluxo utiliza PyAutoGUI para executar ações equivalentes às do usuário
- Localiza o botão Atualizar  
- Clica sobre ele  
- Aguarda o processamento de todas as consultas  
- Captura possíveis mensagens de aviso ou janelas modais  
- Conclui o refresh da base
- Como depende da interface gráfica, usa coordenadas pré-mapeadas da tela

📁 <span style="color: MediumSeaGreen"><b> 3- Publicação automática no workspace corporativo </b></span></p>
- Clique em Publicar  
- Seleção do workspace pré-definido (ex.: GESTÃO DE PESSOAS)  
- Confirmação de Substituir relatório existente  
- Aguardar finalização da publicação
- Essa etapa elimina totalmente o processo manual realizado na interface do Power BI Desktop/Service

🔄 <span style="color: MediumSeaGreen"><b> 4- Sincronização completa do dashboard </b></span></p>
- Ao final da automação, a versão publicada contém:
- Dados recentes  
- Modelos recalculados  
- Medidas atualizadas  
- Tabelas e relacionamentos sincronizados
- O dashboard acessado pelo RH e diretoria passa a refletir imediatamente os dados mais novos de Recrutamento & Seleção

🧾 <span style="color: MediumSeaGreen"><b> 5- Encerramento automático </b></span></p>
- O Power BI é finalizado  
- O script imprime um resumo de execução  
- O processo é encerrado

# Atualização do Power BI - ATESTADOS MÉDICOS E AFASTAMENTOS

```python
pyautogui.PAUSE = 1

# Entrar no Power BI
path_pbi = r"X:\Gestão de Pessoas\Analytics\09 - Power BI\ATESTADOS MÉDICOS E AFASTAMENTOS.pbix"
os.startfile(path_pbi) # Abre o arquivo
time.sleep(30)

# Atualizar Power BI
pyautogui.click(x=753, y=103) # Clicar em Atualizar
time.sleep(60)
pyautogui.click(x=1293, y=98) # Publicar
time.sleep(5)
pyautogui.click(x=863, y=477) # Salvar
time.sleep(5)
pyautogui.press("tab")
pyautogui.press("tab")
pyautogui.press("tab")
pyautogui.press("enter")
time.sleep(1)
pyautogui.press("tab")
pyautogui.press("enter")
time.sleep(7)
pyautogui.click(x=871, y=577) # Substituir
time.sleep(10)
pyautogui.click(x=990, y=585) # Clicar em Entendi
time.sleep(3)
pyautogui.hotkey("Alt", "F4")
time.sleep(2)

print('----------------------------------------------------------------------------------------------------')
print('')
print('   ✅ Power BI atualizado')
print('')
print('----------------------------------------------------------------------------------------------------')
```
