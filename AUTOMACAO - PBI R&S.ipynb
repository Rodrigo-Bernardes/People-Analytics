# Mover arquivo para pasta de Extração

```python
# Caminho completo do arquivo de origem
origem_arquivo = r'C:\Users\rodrigo.bernandes\Downloads\CONTROLE DE VAGAS.xlsb'

# Caminho completo do diretório de destino
destino_diretorio = r'X:\Gestao_de_Pessoas\Analytics\03 - Bases\1. BASES TRATADAS'

# Nome do arquivo a ser movido
nome_arquivo = os.path.basename(origem_arquivo)

# Caminho completo do arquivo no destino
destino_arquivo = os.path.join(destino_diretorio, nome_arquivo)

# Verifica se o arquivo de origem existe
if os.path.exists(origem_arquivo):
    try:
        # Verifica se já existe um arquivo com o mesmo nome no destino
        if os.path.exists(destino_arquivo):
            print(f"O arquivo '{nome_arquivo}' já existe no destino. Removendo...")
            os.remove(destino_arquivo) # Remove o arquivo existente
            print("Arquivo antigo removido com sucesso.")

        # Move o arquivo para o diretório de destino (agora pode ser movido ou sobrescrever se o nome for o mesmo e o antigo foi removido)
        shutil.move(origem_arquivo, destino_diretorio)
        print(f"Arquivo '{nome_arquivo}' movido e substituído com sucesso em '{destino_diretorio}'")

    except FileNotFoundError:
        print(f"Erro: O diretório de destino '{destino_diretorio}' não foi encontrado.")
    except PermissionError:
        print(f"Erro: Permissão negada. Verifique se você tem direitos de escrita em '{destino_diretorio}' ou se o arquivo não está em uso.")
    except Exception as e:
        print(f"Ocorreu um erro inesperado ao mover o arquivo: {e}")
else:
    print(f"Erro: O arquivo de origem '{origem_arquivo}' não foi encontrado.")

print('----------------------------------------------------------------------------------------------------')
print('')
print('   ✅ Arquivo movido')
print('')
print('----------------------------------------------------------------------------------------------------')
```
