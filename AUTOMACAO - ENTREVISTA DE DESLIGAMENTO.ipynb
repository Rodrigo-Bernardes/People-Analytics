```python
time.sleep(5)

class Config:
    """Centraliza todas as configurações do projeto"""
    
    PROCESS_ID = 9
    BASE_PATH = r'X:\Gestão de Pessoas\Analytics\03 - Bases'
    
    PATHS = {
        'processo_registro': rf'{BASE_PATH}\1. BASES TRATADAS\PROCESSOS.xlsx',
        'extracao': r'C:\Users\rodrigo.bernandes\Downloads',
        'movidos': rf'{BASE_PATH}\2. ARQUIVOS MOVIDOS',
        'final_colaborador': rf'{BASE_PATH}\1. BASES TRATADAS\ENTREVISTA DE DESLIGAMENTO COLABORADOR.xlsx',
        'final_gestor': rf'{BASE_PATH}\1. BASES TRATADAS\ENTREVISTA DE DESLIGAMENTO GESTOR.xlsx',
        'depara_colaborador': rf'{BASE_PATH}\1. BASES TRATADAS\DE-PARA ID ENTREVISTA DE DESLIGAMENTO COLABORADOR.xlsx',
        'depara_gestor': rf'{BASE_PATH}\1. BASES TRATADAS\DE-PARA ID ENTREVISTA DE DESLIGAMENTO GESTOR.xlsx',
    }
    
    TIPOS_ENTREVISTA = {
        'colaborador': {
            'arquivo_filtro': 'ENTREVISTA DE DESLIGAMENTO - COLABORADOR',
            'comprimento_filtro': 40,
            'depara_key': 'depara_colaborador',
            'caminho_saida': 'final_colaborador',
            'nome_sheet': 'ENT DESLIGAMENTOS - COLABORADOR',
        },
        'gestor': {
            'arquivo_filtro': 'ENTREVISTA DE DESLIGAMENTO - ANÁLISE DO GESTOR',
            'comprimento_filtro': 46,
            'depara_key': 'depara_gestor',
            'caminho_saida': 'final_gestor',
            'nome_sheet': 'ENT DESLIGAMENTOS - GESTOR',
        }
    }

# ============================================================================
# CONFIGURAÇÃO DE LOGGING
# ============================================================================

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)

# ============================================================================
# FUNÇÕES AUXILIARES - LIMPEZA DE COLUNAS
# ============================================================================

def limpar_nomes_colunas(df):
    """Remove espaços extras e caracteres especiais dos nomes de colunas"""
    df.columns = df.columns.str.strip()
    return df

def encontrar_coluna(df, termo_busca):
    """Encontra coluna por termo parcial (case-insensitive)"""
    termo = termo_busca.lower().strip()
    for col in df.columns:
        if termo in col.lower():
            return col
    return None

def mapear_colunas_entrevista_colaborador(ent_desl_col):
    """Mapeia as colunas reais do DataFrame para nomes padronizados."""
    ent_desl_col = limpar_nomes_colunas(ent_desl_col)
    
    mapeamento = {
        'id': 'ID',
        'hora_inicio': 'Hora de início',
        'hora_conclusao': 'Hora de conclusão',
        'nome_completo': 'Nome completo',
        'principal_motivo': 'principal motivo do seu desligamento',
        'evento_especifico_decisao': 'influenciada por algum evento específico',
        'desc_evento_especifico_decisao': 'nos conte mais',
        'considerou_sair_antes': 'considerado sair antes',
        'clima_organizacional': 'Clima organizacional',
        'relacionamento_colegas': 'Relacionamento com colegas',
        'lideranca': 'Relação com a liderança',
        'reconhecimento_trabalho': 'Reconhecimento pelo trabalho',
        'oportunidade_desenvolvimento': 'Oportunidades de desenvolvimento',
        'beneficios': 'Benefícios oferecidos',
        'pontos_positivos': 'pontos mais positivos de trabalhar',
        'pontos_melhorar': 'pontos mais desafiadores',
        'suporte_atividades': 'suporte suficiente para realizar suas atividades',
        'clareza_lideranca': 'liderança foi clara ao comunicar',
        'feedback_construtivo': 'feedback recebido sobre o seu trabalho',
        'treinamento_adequado': 'treinamento adequado',
        'acao_evitar_saida': 'algo poderia ter sido feito para evitar sua saída',
        'qual_acao_evitar_saida': 'principal fator que influenciaria sua permanência',
        'sugestao_melhoria': 'principais sugestões que você gostaria de deixar',
        'recomendaria': 'recomendaria a empresa como um excelente lugar',
        'satisfacao': 'sua satisfação geral com a empresa',
        'ponto_compartilhar': 'outro ponto que você gostaria de compartilhar',
    }
    
    colunas_encontradas = {}
    colunas_nao_encontradas = []
    
    for col_padrao, termo_busca in mapeamento.items():
        col_real = encontrar_coluna(ent_desl_col, termo_busca)
        if col_real:
            colunas_encontradas[col_real] = col_padrao
            logger.info(f"✓ Colaborador - Encontrada coluna: {col_padrao}")
        else:
            colunas_nao_encontradas.append(col_padrao)
            logger.warning(f"✗ Colaborador - Coluna não encontrada: {col_padrao}")
    
    if colunas_nao_encontradas:
        logger.error(f"Colunas não mapeadas: {colunas_nao_encontradas}")
    
    ent_desl_col = ent_desl_col.rename(columns=colunas_encontradas)
    return ent_desl_col

def mapear_colunas_entrevista_gestor(ent_desl_ges):
    """Mapeia as colunas reais do DataFrame para nomes padronizados."""
    ent_desl_ges = limpar_nomes_colunas(ent_desl_ges)
    
    mapeamento = {
        'id': 'ID',
        'hora_inicio': 'Hora de início',
        'hora_conclusao': 'Hora de conclusão',
        'email_gestor': 'Email',
        'nome_colaborador': 'Nome completo do colaborador',
        'nome_gestor': 'Nome',
        'performance': 'Como você avalia a performance geral do colaborador',
        'proatividade_comprometimento': 'proatividade e comprometimento',
        'comportamento_prejudicial': 'comportamentos que prejudicaram',
        'qual_comportamento_prejudicial': 'Caso tenha respondido "Sim" no item 4',
        'pontos_fortes': 'principais pontos fortes',
        'pontos_melhorar': 'pontos que poderiam ser melhorados',
        'alinhado_cultura': 'alinhado aos valores e à cultura',
        'dificuldade_adaptar_ambiente': 'dificuldade em adaptar-se',
        'principal_motivo': 'principal motivo do desligamento',
        'motivo_evento_especifico': 'evento específico',
        'descricao_motivo_evento_especifico': 'Caso tenha respondido "Sim" no item 11',
        'parecia_comprometido': 'parecia comprometido com suas funções',
        'nivel_engajamento': 'nível de engajamento',
        'relacionamento_colegas_gestores': 'relacionamento do colaborador com colegas',
        'quantidade_feedbacks': 'feedbacks durante o período',
        'quantidade_medidas_disciplinares': 'medidas disciplinares',
        'impacto_saida_equipe': 'saída deste colaborador pode impactar a equipe',
        'impacto_saida_trabalho': 'saída deste colaborador pode impactar o trabalho',
        'acao_reter_talento': 'algo que a empresa poderia ter feito para reter',
        'descricao_acao_reter_talento': 'o que a empresa poderia ter feito ou poderia começar a fazer para ajudar a reter',
        'oportunidades_claras': 'oportunidades claras de crescimento',
        'ponto_compartilhar': 'ponto relevante a ser mencionado',
    }
    
    colunas_encontradas = {}
    colunas_nao_encontradas = []
    
    for col_padrao, termo_busca in mapeamento.items():
        col_real = encontrar_coluna(ent_desl_ges, termo_busca)
        if col_real:
            colunas_encontradas[col_real] = col_padrao
            logger.info(f"✓ Gestor - Encontrada coluna: {col_padrao}")
        else:
            colunas_nao_encontradas.append(col_padrao)
            logger.warning(f"✗ Gestor - Coluna não encontrada: {col_padrao}")
    
    if colunas_nao_encontradas:
        logger.error(f"Colunas não mapeadas: {colunas_nao_encontradas}")
    
    ent_desl_ges = ent_desl_ges.rename(columns=colunas_encontradas)
    return ent_desl_ges

# ============================================================================
# FUNÇÕES AUXILIARES - PROCESSAMENTO GERAL
# ============================================================================

def registrar_etapa(caminho_processo, id_processo, numero_etapa):
    """Registra a etapa do processo no Excel de controle"""
    try:
        wb = load_workbook(caminho_processo)
        ws = wb['REGISTROS']
        ws.append([id_processo, datetime.today(), numero_etapa])
        wb.save(caminho_processo)
        logger.info(f"Etapa {numero_etapa} registrada com sucesso")
    except Exception as e:
        logger.error(f"Erro ao registrar etapa {numero_etapa}: {e}")
        raise

def carregar_depara(caminho_depara):
    """Carrega arquivo de De-PARA"""
    try:
        df = pd.read_excel(caminho_depara)
        logger.info(f"De-PARA carregado: {len(df)} registros")
        return df
    except Exception as e:
        logger.error(f"Erro ao carregar De-PARA de {caminho_depara}: {e}")
        raise

def listar_arquivos_filtrados(diretorio, filtro, comprimento):
    """Lista arquivos do diretório, filtra e ordena usando Pandas"""
    try:
        arquivos = os.listdir(diretorio)
        df = pd.DataFrame(arquivos, columns=['arquivo'])
        
        # Filtra os arquivos que começam com o texto desejado
        df_filtrado = df[df['arquivo'].str.startswith(filtro)].copy()
        
        # Extrai o número entre parênteses para ordenação (ex: "(1)", "(2)")
        # Se não houver número, preenche com 0
        df_filtrado['ordem'] = df_filtrado['arquivo'].str.extract(r'\((\d+)\)').fillna(0).astype(int)
        
        # Ordena e remove a coluna auxiliar
        df_filtrado = df_filtrado.sort_values('ordem').drop(columns=['ordem'])
        
        logger.info(f"Encontrados {len(df_filtrado)} arquivos para filtro: {filtro}")
        return df_filtrado
    except Exception as e:
        logger.error(f"Erro ao listar arquivos filtrados: {e}")
        raise

def criar_tabela_excel(dataframe, caminho_saida, nome_sheet="Dados", nome_tabela="ENTREVISTAS"):
    """Cria arquivo Excel com formatação de tabela"""
    try:
        wb = Workbook()
        ws = wb.active
        ws.title = nome_sheet
        
        for r in dataframe_to_rows(dataframe, index=False, header=True):
            ws.append(r)
        
        tabela = Table(displayName=nome_tabela, ref=ws.dimensions)
        estilo = TableStyleInfo(
            name="TableStyleLight13",
            showFirstColumn=False,
            showLastColumn=False,
            showRowStripes=True,
            showColumnStripes=True
        )
        tabela.tableStyleInfo = estilo
        ws.add_table(tabela)
        
        wb.save(caminho_saida)
        logger.info(f"Arquivo Excel criado: {caminho_saida}")
    except Exception as e:
        logger.error(f"Erro ao criar Excel em {caminho_saida}: {e}")
        raise

def mover_arquivo(origem, destino):
    """Move arquivo de origem para destino"""
    try:
        shutil.move(origem, destino)
        logger.info(f"Arquivo movido: {origem} → {destino}")
    except Exception as e:
        logger.error(f"Erro ao mover arquivo {origem}: {e}")
        raise

# ============================================================================
# FUNÇÕES DE PROCESSAMENTO - ENTREVISTA COLABORADOR
# ============================================================================

def query_processar_entrevista_colaborador(ent_desl_col, depara_id_colaborador):
    """Processa entrevista de desligamento - Colaborador usando Pandas (Substitui DuckDB)"""
    try:
        # Garante que os IDs sejam strings para o merge funcionar perfeitamente
        ent_desl_col['id'] = ent_desl_col['id'].astype(str)
        depara_id_colaborador['id'] = depara_id_colaborador['id'].astype(str)
        
        # Faz o LEFT JOIN
        resultado = pd.merge(
            ent_desl_col,
            depara_id_colaborador[['id', 'registro']],
            on='id',
            how='left'
        )
        
        # Define a ordem exata das colunas desejadas
        colunas_finais = [
            'id', 'registro', 'hora_inicio', 'hora_conclusao', 'nome_completo',
            'principal_motivo', 'evento_especifico_decisao', 'desc_evento_especifico_decisao',
            'considerou_sair_antes', 'clima_organizacional', 'relacionamento_colegas',
            'lideranca', 'reconhecimento_trabalho', 'oportunidade_desenvolvimento',
            'beneficios', 'pontos_positivos', 'pontos_melhorar', 'suporte_atividades',
            'clareza_lideranca', 'feedback_construtivo', 'treinamento_adequado',
            'acao_evitar_saida', 'qual_acao_evitar_saida', 'sugestao_melhoria',
            'recomendaria', 'satisfacao', 'ponto_compartilhar'
        ]
        
        # Filtra apenas as colunas que realmente existem no DataFrame para evitar KeyError
        colunas_presentes = [col for col in colunas_finais if col in resultado.columns]
        resultado = resultado[colunas_presentes]
        
        # IDENTIFICAR REGISTROS SEM CORRESPONDÊNCIA NO DE-PARA
        registros_sem_depara = resultado[resultado['registro'].isna()]
        if len(registros_sem_depara) > 0:
            ids_faltantes = registros_sem_depara['id'].tolist()
            logger.warning(f"⚠️ Encontrados {len(registros_sem_depara)} registros SEM correspondência no De-PARA:")
            logger.warning(f"   IDs faltantes: {ids_faltantes}")
        
        logger.info(f"Processamento colaborador executado com sucesso: {len(resultado)} registros")
        return resultado
    except Exception as e:
        logger.error(f"Erro ao processar colaborador: {e}")
        raise

# ============================================================================
# FUNÇÕES DE PROCESSAMENTO - ENTREVISTA GESTOR
# ============================================================================

def query_processar_entrevista_gestor(ent_desl_ges, depara_id_gestor):
    """Processa entrevista de desligamento - Gestor usando Pandas (Substitui DuckDB)"""
    try:
        # Cria a chave concatenada no formato "ID N" para o cruzamento
        ent_desl_ges['id_concat'] = ent_desl_ges['id'].astype(str) + ' N'
        depara_id_gestor['id'] = depara_id_gestor['id'].astype(str)
        
        # Faz o LEFT JOIN usando a chave concatenada
        resultado = pd.merge(
            ent_desl_ges,
            depara_id_gestor[['id', 'registro']],
            left_on='id_concat',
            right_on='id',
            how='left',
            suffixes=('_orig', '')
        )
        
        # Substitui a coluna 'id' original pela concatenada (como era no SQL)
        resultado['id'] = resultado['id_concat']
        
        # Define a ordem exata das colunas desejadas
        colunas_finais = [
            'id', 'registro', 'hora_inicio', 'hora_conclusao', 'email_gestor', 'nome_gestor',
            'nome_colaborador', 'performance', 'proatividade_comprometimento',
            'comportamento_prejudicial', 'qual_comportamento_prejudicial',
            'pontos_fortes', 'pontos_melhorar', 'alinhado_cultura',
            'dificuldade_adaptar_ambiente', 'principal_motivo',
            'motivo_evento_especifico', 'descricao_motivo_evento_especifico',
            'parecia_comprometido', 'nivel_engajamento', 'relacionamento_colegas_gestores',
            'quantidade_feedbacks', 'quantidade_medidas_disciplinares',
            'impacto_saida_equipe', 'impacto_saida_trabalho', 'acao_reter_talento',
            'descricao_acao_reter_talento', 'oportunidades_claras', 'ponto_compartilhar'
        ]
        
        # Filtra apenas as colunas que realmente existem no DataFrame
        colunas_presentes = [col for col in colunas_finais if col in resultado.columns]
        resultado = resultado[colunas_presentes]
        
        # IDENTIFICAR REGISTROS SEM CORRESPONDÊNCIA NO DE-PARA
        registros_sem_depara = resultado[resultado['registro'].isna()]
        if len(registros_sem_depara) > 0:
            ids_faltantes = registros_sem_depara['id'].tolist()
            logger.warning(f"⚠️ Encontrados {len(registros_sem_depara)} registros SEM correspondência no De-PARA:")
            logger.warning(f"   IDs faltantes: {ids_faltantes}")
        
        logger.info(f"Processamento gestor executado com sucesso: {len(resultado)} registros")
        return resultado
    except Exception as e:
        logger.error(f"Erro ao processar gestor: {e}")
        raise

# ============================================================================
# FUNÇÃO PRINCIPAL DE PROCESSAMENTO
# ============================================================================

def processar_entrevistas():
    """Função principal que orquestra todo o processo"""
    
    tempo_inicio = datetime.today()
    logger.info("=" * 80)
    logger.info("INICIANDO PROCESSAMENTO DE ENTREVISTAS DE DESLIGAMENTO")
    logger.info("=" * 80)
    
    df_consolidado_colaborador = pd.DataFrame()
    df_consolidado_gestor = pd.DataFrame()
    
    try:
        # Etapa 0
        registrar_etapa(Config.PATHS['processo_registro'], Config.PROCESS_ID, 0)
        
        # Etapa 1
        logger.info("Carregando arquivos De-PARA...")
        depara_colaborador = carregar_depara(Config.PATHS['depara_colaborador'])
        depara_gestor = carregar_depara(Config.PATHS['depara_gestor'])
        registrar_etapa(Config.PATHS['processo_registro'], Config.PROCESS_ID, 1)
        
        # Etapa 2
        logger.info("Filtrando arquivos de entrada...")
        config_col = Config.TIPOS_ENTREVISTA['colaborador']
        config_ges = Config.TIPOS_ENTREVISTA['gestor']
        
        df_arquivos_colaborador = listar_arquivos_filtrados(
            Config.PATHS['extracao'],
            config_col['arquivo_filtro'],
            config_col['comprimento_filtro']
        )
        
        df_arquivos_gestor = listar_arquivos_filtrados(
            Config.PATHS['extracao'],
            config_ges['arquivo_filtro'],
            config_ges['comprimento_filtro']
        )
        registrar_etapa(Config.PATHS['processo_registro'], Config.PROCESS_ID, 2)
        
        # Etapa 3: COLABORADOR
        logger.info("Processando entrevistas de COLABORADOR...")
        for arquivo in df_arquivos_colaborador['arquivo'].tolist():
            try:
                caminho_arquivo = os.path.join(Config.PATHS['extracao'], arquivo)
                ent_desl_col = pd.read_excel(caminho_arquivo)
                logger.info(f"Arquivo {arquivo} contém {len(ent_desl_col)} registros brutos")
                
                ent_desl_col = mapear_colunas_entrevista_colaborador(ent_desl_col)
                ent_desl_col = query_processar_entrevista_colaborador(ent_desl_col, depara_colaborador)
                
                df_consolidado_colaborador = pd.concat([df_consolidado_colaborador, ent_desl_col], ignore_index=True)
                mover_arquivo(caminho_arquivo, os.path.join(Config.PATHS['movidos'], arquivo))
                logger.info(f"✓ Arquivo processado: {arquivo} ({len(ent_desl_col)} registros incluídos no consolidado)")
            except Exception as e:
                logger.error(f"Erro ao processar arquivo {arquivo}: {e}")
                continue
        
        if not df_consolidado_colaborador.empty:
            logger.info(f"Salvando consolidado de COLABORADOR: {len(df_consolidado_colaborador)} registros")
            criar_tabela_excel(df_consolidado_colaborador, Config.PATHS['final_colaborador'], config_col['nome_sheet'])
        else:
            logger.warning("Nenhum registro de COLABORADOR para salvar")
        
        registrar_etapa(Config.PATHS['processo_registro'], Config.PROCESS_ID, 3)
        
        # Etapa 4: GESTOR
        logger.info("Processando entrevistas de GESTOR...")
        for arquivo in df_arquivos_gestor['arquivo'].tolist():
            try:
                caminho_arquivo = os.path.join(Config.PATHS['extracao'], arquivo)
                ent_desl_ges = pd.read_excel(caminho_arquivo)
                logger.info(f"Arquivo {arquivo} contém {len(ent_desl_ges)} registros brutos")
                
                ent_desl_ges = mapear_colunas_entrevista_gestor(ent_desl_ges)
                ent_desl_ges = query_processar_entrevista_gestor(ent_desl_ges, depara_gestor)
                
                df_consolidado_gestor = pd.concat([df_consolidado_gestor, ent_desl_ges], ignore_index=True)
                mover_arquivo(caminho_arquivo, os.path.join(Config.PATHS['movidos'], arquivo))
                logger.info(f"✓ Arquivo processado: {arquivo} ({len(ent_desl_ges)} registros incluídos no consolidado)")
            except Exception as e:
                logger.error(f"Erro ao processar arquivo {arquivo}: {e}")
                continue
        
        if not df_consolidado_gestor.empty:
            logger.info(f"Salvando consolidado de GESTOR: {len(df_consolidado_gestor)} registros")
            criar_tabela_excel(df_consolidado_gestor, Config.PATHS['final_gestor'], config_ges['nome_sheet'])
        else:
            logger.warning("Nenhum registro de GESTOR para salvar")
        
        registrar_etapa(Config.PATHS['processo_registro'], Config.PROCESS_ID, 4)
        
        # Resumo final
        tempo_fim = datetime.today()
        tempo_total = tempo_fim - tempo_inicio
        
        logger.info("=" * 80)
        logger.info("PROCESSO FINALIZADO COM SUCESSO")
        logger.info(f"Total Colaborador: {len(df_consolidado_colaborador)} registros")
        logger.info(f"Total Gestor: {len(df_consolidado_gestor)} registros")
        logger.info(f"Tempo total de execução: {tempo_total}")
        logger.info("=" * 80)
        
    except Exception as e:
        logger.critical(f"Erro crítico no processo: {e}")
        raise

if __name__ == "__main__":
    processar_entrevistas()
```
